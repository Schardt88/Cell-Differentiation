```python
from Organoid2D import Organoid
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

org = Organoid()
org.evolution(T=0.001, file = 'organoid_mid.csv', mode='transcription')
org.cellPlot()
```

```python
org.mesh = {}

# Number of lines in that dimension which cut through the organoid
Nx = 1
Ny = 1

x = np.linspace(np.min(org.xy[:,0]) - org.r_max/2, np.max(org.xy[:,0]) + org.r_max/2, Nx+2)
y = np.linspace(np.min(org.xy[:,1]) - org.r_max/2, np.max(org.xy[:,1]) + org.r_max/2, Ny+2)

org.N = np.zeros(org.nofCells)
for i in range(Nx+1):
    for j in range(Ny+1):
        if (i+j)%2 == 0:
            org.N[(x[i] <= org.xy[:,0]) & (org.xy[:,0] < x[i+1]) & (y[j] <= org.xy[:,1]) & (org.xy[:,1] < y[j+1])] = 1   

plt.figure(figsize=[6.4, 4.8])
plt.rc('font', size=14)
org.G = 1 - org.N
org.cellPlot(org.N, size=0)
plt.savefig('test.png', transparent=True)

im = plt.imread('test.png')

fig=plt.figure()
org.pcf(font_size = 14*3/2)
newax = fig.add_axes([.64,.64,0.5,0.5], zorder=1)
newax.imshow(im)
newax.axis('off')

org.moran()
print(org.Morans_I)
```

```python
norm = (org.xy[:,0]**2 + org.xy[:,1]**2)**0.5

Nr = 2
r1 = (np.max(norm))/Nr**0.5
r = np.empty(Nr+2)
for i in range(Nr+2):
    r[i] = (i)**0.5*r1
    
print(r)
org.N = np.zeros(org.nofCells)
for i in range(Nr+1):
    if i%2 == 0:
        org.N[(r[i] < norm) & (norm <= r[i+1])] = 1

org.cellPlot(org.N)
print(len(org.N[org.N==1]),len(org.N[org.N==0]))
```
