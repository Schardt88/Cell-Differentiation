# Organoids

```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import numpy as np

C = pd.read_csv('includingSurfaceDistance/extendedRawDataICMOrganoids.csv')

Organoids = max(C['OrganoidID'])
ID = 50
Cindex = []
Cells = []

def fate(x,y):
    val = np.array([0]*len(y))
    p = 10
    for i in range(len(y)):
        if x[i]/max(x) >= y[i]/max(y):
            val[i] = 1
        #if y[i]/max(y) >= 3*x[i]/max(x):
        #    val[i] = -1
            
    return val

for i in range(len(C)):
    if C['OrganoidID'][i] == ID:
        Cindex.append(i)

x = np.array(C.loc[Cindex,'CentroidX'])
y = np.array(C.loc[Cindex,'CentroidY'])
z = np.array(C.loc[Cindex,'CentroidZ'])
N = np.array(C.loc[Cindex,'Nanog-Avg'])
G = np.array(C.loc[Cindex,'Gata6-Avg'])
Nmax = max(C.loc[Cindex,'Nanog-Avg'])
Gmax = max(C.loc[Cindex,'Gata6-Avg'])
Nmin = min(C.loc[Cindex,'Nanog-Avg'])
Gmin = min(C.loc[Cindex,'Gata6-Avg'])
Centre = np.array([sum(x)/len(x), sum(y)/len(x), sum(z)/len(x)])

pos = np.array([[0]*3]*len(x), dtype = 'float')
Cdist = np.array([0]*len(x), dtype = 'float')
for i in range(len(x)):
    pos[i][0] = x[i]
    pos[i][1] = y[i]
    pos[i][2] = z[i]
    Cdist[i] = np.linalg.norm(pos[i]-Centre)

plt.figure(figsize=(20,10))
ax = plt.axes(projection='3d')
ax.scatter3D(x, y, z, c = N/Nmax - G/Gmax, s=30, cmap='cool');

plt.figure(figsize=(20,5))
plt.subplot(1,3,1)
plt.scatter(x, y, c = N/Nmax - G/Gmax, s=30, cmap='cool');
plt.xlabel('x')
plt.ylabel('y')
plt.subplot(1,3,2)
plt.scatter(y, z, c = N/Nmax - G/Gmax, s=30, cmap='cool');
plt.xlabel('y')
plt.ylabel('z')
plt.subplot(1,3,3)
plt.scatter(z, x, c = N/Nmax - G/Gmax, s=30, cmap='cool');
plt.xlabel('z')
plt.ylabel('x')

plt.figure(figsize=(20,5))
plt.subplot(1,3,1)
plt.scatter(Cdist, N);
plt.xlabel('distance to centre of mass')
plt.ylabel('Nanog')
plt.subplot(1,3,2)
plt.scatter(Cdist, G);
plt.xlabel('distance to centre of mass')
plt.ylabel('Gata6')
plt.subplot(1,3,3)
plt.scatter(Cdist,  N/max(N) - G/max(G));
plt.xlabel('distance to centre of mass')
plt.ylabel('Nanog - Gata6 (normalized)')

print(np.corrcoef(Cdist, fate(N,G)))
```

# Energy Functional

```python
from math import exp
import scipy.optimize as opt


pos = np.array([[0]*3]*len(x), dtype = 'float')
for i in range(len(x)):
    pos[i][0] = x[i]
    pos[i][1] = y[i]
    pos[i][2] = z[i]

k = -1
N_tot = 0
G_tot = 0
dist = np.array([[0]*len(x)]*len(x), dtype = 'float')
for i in range(len(x)):
    for j in range(len(x)):
        if i != j:
            dist[i][j] = np.linalg.norm(pos[i]-pos[j])
    
    N_tot += N[i]
    G_tot += G[i]
    
dist = dist/np.amax(dist)

def Energy(NG):
    val = 0
    for i in range(len(x)):
        val_i = 0
        for j in range(len(x)):
            if i != j:
                Ni = NG[i]
                Nj = NG[j]
                Gi = NG[i+len(x)]
                Gj = NG[j+len(x)]
                val_i += dist[i][j]**k*Ni*Gj
            
        val += val_i
    
    return val

def dEnergy(NG):
    val = np.array([0]*len(NG), dtype='float')
    for i in range(len(x)):
        val_Ni = 0
        val_Gi = 0
        for j in range(len(x)):
            if i != j:
                Nj = NG[j]
                Gj = NG[j+len(x)]
                val_Ni += dist[i][j]**k*Gj
                val_Gi += dist[i][j]**k*Nj
            
        val[i] = val_Ni
        val[i+len(x)] = val_Gi
    
    return val

I = np.identity(len(x))
O = np.array([[0]*len(x)]*len(x))
C = np.block([[I, O], [O, I], [I, I]])
lower_bound = np.block([np.array([Nmin]*len(x)), np.array([Gmin]*len(x)), np.array([Nmin + Gmin]*len(x))])
upper_bound = np.block([np.array([Nmax]*len(x)), np.array([Gmax]*len(x)), np.array([Nmax + Gmax]*len(x))])

linear_constraint = opt.LinearConstraint(C, lower_bound, upper_bound)
NG0 = np.block([N, G])

#res = opt.minimize(Energy, NG0, method='trust-constr', jac = '2-point', hess = BFGS(),
#                  constraints=linear_constraint, options = {'maxiter': 3, 'disp': True})

#print(res)
#N_sol = np.array([0]*len(x), dtype = 'float')
#G_sol = np.array([0]*len(x), dtype = 'float')
#for i in range(len(x)):
#    N_sol[i] = Sol[i]
#    G_sol[i] = Sol[i+len(x)]
```

```python
import random as rd
from scipy.integrate import solve_ivp
############# INITIALIZE MODEL #############
# Time integration related parameters
CellPop = len(x)
T = 100
Steps = 30
t = np.linspace(0, T, Steps)

# Inhibition Intensity from NANOG to GATA6
a = 0.001

# Inhibition Intensity from GATA6 to NANOG
b = 0.01

# Activation Intensity from lateral NANOG to GATA6
c = 0.04

# Exponents
n = 2

# Inhibition / Activation Functions
def f(a,x):
    f = a/(a+x**n)
    return f

def g(b,x):
    g = b/(b+x**n)
    return g

def h(c,x):
    h = x**n/(c+x**n)
    return h

def conv_G(x,i):
    val = 0
    for j in range(CellPop):
        if i == j:
            val=val
        else:
            R = 0.2 #4*x[j+CellPop]**2 #np.exp(x[j+CellPop]/(1.5-x[j+CellPop]))-1
            if dist[i][j] <= R:
                val +=x[j+CellPop]/dist[i][j]/CellPop
    return val

def rhs(t, x):
    val = [0]*2*CellPop
    for i in range(CellPop):
        val[i] = f(a,x[i+CellPop]) + h(c,conv_G(x,i)) - x[i]
        val[i+CellPop] = g(b,x[i]) - x[i+CellPop]

    return val

xInit = [0]*2*CellPop
for i in range(CellPop):
    xInit[i] = 0.3# + rd.uniform(-0.005,0.005)
    xInit[i+CellPop] = g(b, xInit[i])

print('Parameters\n a =',a,'\n b =',b,'\n c =',c,'\n n =',n,'\n T =',T,'\n TimeSteps =',Steps)

sol = solve_ivp(rhs, [0, T], xInit, t_eval = t, method = 'RK23')


N_ode = [0]*CellPop
G_ode = [0]*CellPop

for i in range(CellPop):
    N_ode[i] = sol.y[i,-1]
    G_ode[i] = sol.y[i+CellPop,-1]
    
plt.figure(figsize=(15,10))
plt.subplot(2, 2, 1)
for i in range(CellPop):
    plt.plot(t, sol.y[i,:])

plt.title('NANOG')
plt.xlabel('Time')
plt.ylabel('Concentrations')

plt.subplot(2, 2, 2)
for i in range(CellPop):
    plt.plot(t, sol.y[i+CellPop,:])

plt.title('GATA6')
plt.xlabel('Time')
```
